## Goal of the Project

**To recognize fraudulent credit card transactions so that the customers of credit card companies are not charged for items that they did not purchase.**

## Step 1: Importing Important Libraries

```python
# Let's import basic libraries first.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Loading the Data

```python
# Loading the data from the local drive
df=pd.read_csv("creditcard.csv")
df.head()
```

## Step 4: Modelling

```python
# Let's split the data into features and target
X= df.drop("Class", axis=1)
y= df["Class"]
```

```python
print(X.shape)
print(y.shape)
```

```python
# Now let's spilt the data into training and test

from sklearn.model_selection import train_test_split
np.random.seed(42)
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2)
```

### Trying Different Models

#### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
np.random.seed(42)
L_model= LogisticRegression()
L_model.fit(X_train, y_train)
L_model.score(X_test, y_test)
```

#### 2. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
R_model = RandomForestClassifier()
R_model.fit(X_train, y_train)
R_model.score(X_test, y_test)
```

```python
ypred = R_model.predict(X_test)
ypred
```

```python
from sklearn.metrics import classification_report, accuracy_score  
from sklearn.metrics import precision_score, recall_score 
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix 

accuracy= accuracy_score(y_test, ypred)
accuracy
```

```python
precision= precision_score(y_test, ypred)
precision
```

```python
rec = recall_score(y_test, ypred)
rec
```

```python
f1 = f1_score(y_test, ypred) 
f1
```

```python

# printing the confusion matrix 
LABELS = ['Normal', 'Fraud'] 
confusion_mat = confusion_matrix(y_test, ypred) 
plt.figure(figsize =(6, 4)) 
sns.heatmap(confusion_mat, xticklabels = LABELS,  
            yticklabels = LABELS, annot = True, fmt ="d"); 
plt.title("Confusion matrix") 
plt.ylabel('True class') 
plt.xlabel('Predicted class') 
plt.show() 
```
